## Visualization & Explanatory Analysis

### Pertanyaan 1: pada jam berapa yang paling banyak dan paling sedikit disewa?

```python
# Pertanyaan 1: Jam dengan penyewaan terbanyak dan tersedikit
sum_order_items_df = hour_df.groupby("hr").cnt.sum().reset_index()

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Menyimpan data top 5 terbanyak dan tersedikit ke dalam variabel
top_5_hours = sum_order_items_df.sort_values(by="cnt", ascending=False).head(5)
bottom_5_hours = sum_order_items_df.sort_values(by="cnt", ascending=True).head(5)

# Jam Terbanyak (Tambahkan parameter 'order')
sns.barplot(x="hr", y="cnt", data=top_5_hours, palette=colors, ax=ax[0], order=top_5_hours["hr"])
ax[0].set_ylabel(None)
ax[0].set_xlabel("Jam (hr)", fontsize=12)
ax[0].set_title("Jam Penyewaan Sepeda Terbanyak", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=12)

# Jam Tersedikit (Tambahkan parameter 'order')
sns.barplot(x="hr", y="cnt", data=bottom_5_hours, palette=colors, ax=ax[1], order=bottom_5_hours["hr"])
ax[1].set_ylabel(None)
ax[1].set_xlabel("Jam (hr)", fontsize=12)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Jam Penyewaan Sepeda Tersedikit", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Pola Penyewaan Sepeda Berdasarkan Jam", fontsize=20)
plt.show()
```

**Insight:**
- Jam penyewaan terbanyak terjadi pada jam 17.00 dengan 300.000 lebih penyewa.
- Jam penyewaan tersedikit terjadi pada jam 04.00  dengan hanya kurang dari 5000 penyewa.

### Pertanyaan 2: pada musim apa penyewaan sepeda paling banyak?

```python
# Mapping nilai season biar lebih mudah dibaca (1:Springer, 2:Summer, 3:Fall, 4:Winter)
season_mapping = {1: 'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
day_df['season_name'] = day_df['season'].map(season_mapping)

season_rentals = day_df.groupby('season_name')['cnt'].sum().reset_index()

plt.figure(figsize=(10, 6))
colors_season = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"] # Asumsi Fall tertinggi

sns.barplot(
    x="season_name", 
    y="cnt",
    data=season_rentals.sort_values(by="cnt", ascending=False),
    palette=colors_season
)
plt.title("Total Penyewaan Sepeda Berdasarkan Musim", fontsize=15)
plt.xlabel("Musim", fontsize=12)
plt.ylabel("Total Penyewaan (cnt)", fontsize=12)
plt.show()
```

**Insight:**
- Total penyewaan sepeda terbesar terjadi pada Musim GUGUR serta tersedikit pada Musim SEMI

### Pertanyaan 3: Bagaimana tren performa total penyewaan sepeda perusahaan per kuartal secara spesifik pada periode tahun 2011 hingga 2012?

```python
plt.figure(figsize=(14, 5))

#mengelompokkan data perkuartal
quarterly_counts = (
    day_df
    .set_index('dteday')
    .resample('Qs')['cnt']
    .sum()
)

# Menambahkan marker
plt.plot(quarterly_counts.index, quarterly_counts.values, marker='o')
plt.scatter(quarterly_counts.index, quarterly_counts.values, s=60)

plt.xlabel('Quartal')
plt.ylabel('Total Penyewaan')
plt.title('Total Penyewaan Sepeda per Quartal (2011 - 2012)')

# Menambahkan grid agar lebih mudah melihat perpotongan nilai
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

**Insight:**
- Peningkatan penyewaan sepeda secara spesifik terjadi pada kuartal pertama hingga kuartal ketiga pada setiap tahunnya
- serta terjadi sedikit penurunan penyewaan sepeda pada kuartal keempat disetiap tahunnya 
- Dari tabel diatas kita dapat mengetahui bahwa pada quartal ke-3 di setiap tahunnya menjadi puncak penyewaan.

### Pertanyaan 4: Bagaimana perbandingan total penggunaan antara pengguna registered dan casual?

```python
# Menghitung total masing-masing pengguna
user_total = hour_df[['registered', 'casual']].sum()

plt.figure(figsize=(8, 6))

# Membuat Pie Chart
plt.pie(
    user_total, 
    labels=['Registered\n(' + str(format(user_total['registered'], ',.0f')) + ')', 
            'Casual\n(' + str(format(user_total['casual'], ',.0f')) + ')'], 
    autopct='%1.1f%%', 
    colors=['#72BCD4', '#FFA07A'], 
    explode=(0.1, 0), # Memberikan efek 'potongan keluar' pada bagian Registered
    startangle=90,
    shadow=True,
    textprops={'fontsize': 12}
)

plt.title('Perbandingan Total Pengguna: Registered vs Casual', fontsize=15)
plt.show()
```

**Insight:**
- Perbandingan pengguna menunjukkan ketimpangan yang sangat signifikan, di mana 81.2% dari total keseluruhan penyewaan. Sementara itu, pengguna biasa (Casual) hanya menyumbang 18.8%.

### Pertanyaan 5: Bagaimana pertumbuhan pengguna registered dan casual dari tahun ke tahun?

```python
# Menyiapkan data
yearly_users = hour_df.groupby('yr').agg({
    'registered': 'sum',
    'casual': 'sum'
})

# Mengubah index 0 dan 1 menjadi tahun asli biar mudah dibaca
yearly_users.index = ['2011', '2012']

# Membuat Grouped Bar Chart
ax = yearly_users.plot(kind='bar', figsize=(10, 6), color=['#72BCD4', '#FFA07A'], edgecolor='black', width=0.6)

plt.title('Pertumbuhan Pengguna (Registered vs Casual) Tahun 2011 - 2012', fontsize=15)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Total Penyewaan', fontsize=12)
plt.xticks(rotation=0) # Biar teks tahunnya horizontal

# Menambahkan angka detail di atas setiap batang grafik (biar dapat bintang 5!)
for p in ax.patches:
    ax.annotate(format(p.get_height(), ',.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points',
                fontsize=10)

plt.legend(title='Tipe Pengguna', labels=['Registered', 'Casual'])
plt.tight_layout()
plt.show()
```

**Insight:**
- Berdasarkan grafik pertumbuhan pengguna, terlihat lonjakan penyewaan yang sangat signifikan dari tahun 2011 ke 2012 pada kedua kelompok pengguna. Pengguna 'Registered' mendominasi total penyewaan secara keseluruhan

## Analisis Lanjutan (Opsional)

```python
# Melakukan Binning / Manual Grouping pada jam penyewaan
def categorize_time(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

hour_df['time_category'] = hour_df['hr'].apply(categorize_time)

# Visualisasi hasil clustering
time_category_df = hour_df.groupby('time_category')['cnt'].sum().reset_index()
time_category_df['time_category'] = pd.Categorical(time_category_df['time_category'], categories=['Morning', 'Afternoon', 'Evening', 'Night'], ordered=True)

plt.figure(figsize=(10, 5))
sns.barplot(x='time_category', y='cnt', data=time_category_df.sort_values('time_category'), palette="viridis")
plt.title('Jumlah Penyewaan Berdasarkan Kelompok Waktu (Clustering)', fontsize=15)
plt.xlabel('Kelompok Waktu', fontsize=12)
plt.ylabel('Total Penyewaan', fontsize=12)
plt.show()
```

**Insight:**
- Melalui teknik binning, kita dapat melihat bahwa mayoritas penyewaan terjadi pada kelompok waktu Evening (17:00 - 20:59).
